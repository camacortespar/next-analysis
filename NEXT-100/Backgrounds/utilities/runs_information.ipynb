Make it a dataframe

```python
all_runs = []
# Loop through the entire data structure and flatten it
for period_name, *configs in RUNS_INFO:
    for config_name, runs_dict in configs:
        for run_number, stats in runs_dict.items():
            all_runs.append({
                "run_number": run_number,
                "duration":   stats["duration"],
                "OK":         stats["OK"],
                "LOST":       stats["LOST"],
                "period":     period_name,
                "condition":  config_name,
            })

# Dataframe
MASTER_DF = pd.DataFrame(all_runs).set_index("run_number")
```

Store it in a .csv file

```python
OUTPUT_FILENAME = 'runs_information.csv'
```

```python
MASTER_DF.to_csv(OUTPUT_FILENAME)
```
